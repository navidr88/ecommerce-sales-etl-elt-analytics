# E-commerce Sales ETL/ELT Analytics Project

## Project Overview
This project demonstrates an end-to-end analytics workflow using an e-commerce sales dataset. The pipeline loads raw CSV data into MySQL, applies data quality validation, creates staging and analytical tables, and prepares the data for business reporting.

## Tools Used
- Python
- pandas
- MySQL
- SQL
- MySQL Workbench
- Power BI
- Git / GitHub

## Data Pipeline

CSV Dataset  
↓  
Python Load  
↓  
MySQL Raw Table  
↓  
Staging Layer  
↓  
Data Quality Validation  
↓  
Dimension and Fact Tables  
↓  
Analytics Queries  
↓  
Power BI Dashboard

## Database Tables

- `raw_orders`
- `stg_valid_orders`
- `dim_customers`
- `dim_products`
- `fact_orders`

## Data Quality Findings

The dataset contained invalid date sequences where:

- `ship_date < order_date`
- `delivery_date < ship_date`

Only valid records were kept in the staging layer for analytical modeling.

## Key Analytics

- Revenue by category
- Top products by revenue
- Top customers by revenue
- Payment method distribution
- Monthly revenue trend
